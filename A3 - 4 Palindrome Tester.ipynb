```python
def palindrome_tester():
    s = input('Enter a word, date, statement or sentence to see if it is a palindrome or not: ')

    n_l  = '0123456789'                             # [n]umber [l]ist    (basic building blocks of Decimal Number System)
    a_l = 'qwertyuiopasdfghjklzxcvbnm'              # [a]lphabet [l]ist  (list of all alphabets)



    # Converting the input string into a list of only relevant alphabets and numbers
    s_list = []                                     # [s]tring list
    for i in s:
        if i in n_l or i.lower() in a_l:
            s_list.append(i.lower())
    print('\nThe list of relevant alphabets and numbers is: \n',s_list)



    # Breaking the input into left and right lists
    l_list = []                                    # [l]eft  list 
    r_list = []                                    # [r]ight list 

    c = 1
    a = len(s_list) // 2                             # helpful in dividing the input string into two

    for i in s_list:
        if c <= a:
            l_list.append(i)
        else:
            r_list.append(i)
        
        c += 1
    print('\nThe left  list is: ', l_list)
    print('The right list is: ', r_list)



    # Deleting the first entry of right list if the total number of relevant alphabets/numbers is odd 
    if len(s_list)%2 != 0:
        del r_list[0]
        print('\nSince the total number of relevant alphabets/numbers was odd, so, the new truncated right list is: ',r_list)




    r_list.reverse()                                 # reverses the right_list in order to make the following comparison easy
    print('\nThe reversed right list is: ',r_list)
    print('')

    if l_list == r_list:
        print('\nSince, right and left lists are equal, therefore, THIS IS A PALINDROME')
    else:
        print('\nSince, right and left lists are not equal, therefore, THIS IS NOT A PALINDROME')
    
    print('\n\n\n\n')
    return



palindrome_tester()
palindrome_tester()
palindrome_tester()
```
