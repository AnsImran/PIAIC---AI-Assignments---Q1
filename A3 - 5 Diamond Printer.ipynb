# First Diamond

```python
d_width = int(input('Enter the Max Width of the diamond, the length will be adjusted accordingly: '))         #[d]iamond [w]idth

if d_width%2 == 0:                 # converting the width into an odd number, if it is already not an odd number
    d_width = d_width + 1          # it is necessary to maintain perfect shape of diamond

char = input('Enter the character whose diamond you want to print: ')





'''
This block prints:

       The upper portion of the diamond:           O
                                                  O O
                                                 O   O
                                                O     O                                                                     
'''

c_u_l = (d_width + 1) // 2        #[c]ounter [u]pper [l]eft
c_u_r = 1                         #[c]ounter [u]pper [r]ight


print(f'{str(char).rjust(c_u_l)}')

for i in range(((d_width + 1) // 2) - 1):

    print(f'{str(char).rjust(c_u_l - 1)} {str(char).rjust(c_u_r)}')
    c_u_l = c_u_l - 1
    c_u_r = c_u_r + 2





'''
This block prints:

       The lower portion of the diamond:        O     O
                                                 O   O
                                                  O O
                                                   O                                                                    
'''

c_l_l = 2                       #[c]ounter [l]ower [l]eft
c_l_r = d_width - 4             #[c]ounter [l]ower [r]ight


for i in range(((d_width + 1) // 2) - 2):
    
    print(f'{str(char).rjust(c_l_l)} {str(char).rjust(c_l_r)}')
    c_l_l = c_l_l + 1
    c_l_r = c_l_r - 2

print(f'{str(char).rjust((d_width + 1) // 2)}')
```

# Second Diamond

```python
d_width = int(input('Enter a number:    '))         #[d]iamond [w]idth
print('\n\n')






'''
This block prints:

       The upper portion of the diamond:           O
                                                  O O
                                                 O   O
                                                O     O                                                                     
'''

c_u_l   = d_width                          #[c]ounter [u]pper [l]eft

c_u_r   = 0                                #[c]ounter [u]pper [r]ight

n_p_u_r = d_width                          #[n]umber [p]rinter [u]pper [l]eft


for i in range(d_width + 1):
    
    print(f"{str(c_u_l).rjust(c_u_l+1)}{str(n_p_u_r).rjust(c_u_r + 1)}")
    
    c_u_l   = c_u_l   - 1
    c_u_r   = c_u_r   + 2
    n_p_u_r = n_p_u_r - 1






'''
This block prints:

       The lower portion of the diamond:        O     O
                                                 O   O
                                                  O O
                                                   O                                                                    
'''

c_l_l   = 0                                   #[c]ounter [lower] [l]eft

c_l_r   = d_width*2                           #[c]ounter [l]ower [r]ight

n_p_l_r   = 0                                 #[n]umber [p]rinter [l]ower [r]ight


for i in range(d_width + 1):
    
    print(f"{str(c_l_l).rjust(c_l_l+1)}{str(n_p_l_r).rjust(c_l_r+1)}")
    
    c_l_l   = c_l_l   + 1
    c_l_r   = c_l_r   - 2
    n_p_l_r = n_p_l_r + 1
```
