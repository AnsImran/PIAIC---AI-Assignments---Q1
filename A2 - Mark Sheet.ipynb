```python
import numpy as np



''' This block will take as input initial details. 
'''
Number_Of_Subjects = int(input("Enter the no. of subjects, you want to add in the Marks Sheet: "))
print("\n \n")
    
Marks_Sheet = np.zeros((Number_Of_Subjects,2))
    
Names_Of_Subjects = []



''' This Block will take as input all of the remaining details. It'll store all of the remaining data, related to the table. 
'''
for i in range (Number_Of_Subjects):
    
    Name_Of_Subject = input(f" Enter the name of subject #{i+1}: ")
    Names_Of_Subjects.append(Name_Of_Subject)
    print("\n")
    
    Max_Marks = int(input(f" Enter maximum attainable marks in {Name_Of_Subject}: "))
    Marks_Sheet[i][0] = Max_Marks
    print("\n")
    
    Obtained_Marks = int(input(f" Enter obtained marks in {Name_Of_Subject}: "))
    Marks_Sheet[i][1] = Obtained_Marks
    print("\n \n ")

Total_Max_Marks = np.sum(Marks_Sheet.T[0])

Total_Obtained_Marks = np.sum(Marks_Sheet.T[1])

Obtained_Percentage = (Total_Obtained_Marks / Total_Max_Marks)*100
    
    

    
''' This Block will print all of the details in the form of a well indented Marks Sheet Table. 
'''    
print("\nSubject\t  \t\t Max Marks \t  Obtained marks\n")

for i in range (Number_Of_Subjects):
        
    d = ' '
    c = 11 - len(Names_Of_Subjects[i])
    for j in range(c-1):                               # it helps in proper indentation of the printed table
        d = d + ' '
    if len(Names_Of_Subjects[i]) <= 11:
        Names_Of_Subjects[i] = Names_Of_Subjects[i] + d
        
    print(f"{Names_Of_Subjects[i]}\t   \t {Marks_Sheet[i][0]} \t\t  {Marks_Sheet[i][1]}")

print(f"\nTotal\t  \t\t {Total_Max_Marks} \t\t  {Total_Obtained_Marks}")

print(f"\nObtained Percentage:\t\t\t {Obtained_Percentage: 0.3}% ")
```
