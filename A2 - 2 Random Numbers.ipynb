```python
import numpy as np
from numpy import random
from numpy.random import randint


'''
Generating 100 random numbers between 10 and 90 and storing them in a separate array
'''
q     = int(input("Enter the quantity of random integers that you want to generate: "))
min_n = int(input("Enter the smallest number of the range: "))
max_n = int(input("Enter the biggest  number of the range: "))
print('\n')

# Using for loop
list_q_for = []

for i in range(q):
    list_q_for.append( random.randint(min_n,max_n) )

list_q_for_2 = list_q_for              #copying by value in a separate 'list'
print(f'The list of {q} random integers b/w {min_n}-{max_n} generated using for loop is: \n\n{list_q_for}')



# Using numpy's randint() function
list_q_np   = randint(min_n,max_n, q)

list_q_np_2 = list_q_np                #copying by value in a separate 'array'
print(f"\n\nThe 1D array of {q} random integers b/w {min_n}-{max_n} generated using numpy's randint() is: \n\n{ list_q_np_2 }")





'''
Finding the Sum and Average
'''
# Using for loop
Sum_for = 0
for i in range(q):
    Sum_for = Sum_for + list_q_for[i]

Average_for = Sum_for / len(list_q_for)

print('\n\n\nThe Sum calculated using for loop  is: ', Sum_for,' and the Average is: ', Average_for)


# Using Sum function
Sum_f     = np.sum(list_q_for)
Average_f = np.average(list_q_for)
print('\nThe Sum calculated using numpy.sum is: ', Sum_f,' and the Average is: ', Average_f)





'''
Finding the Maximum Number
'''
# Using for loop
a = min_n
for i in range(q):
    if list_q_for[i] >= a:
        a = list_q_for[i]
print(f'\n\n\nThe biggest number in the random list, found using for loop is: {a}')


# Using np.max
print(f'\nThe biggest number in the random list, found using  np.max  is: {np.max(list_q_for)}')





'''
Finding the Minimum Number
'''
# Using for loop
b = max_n
for i in range(q):
    if list_q_for[i] <= b:
        b = list_q_for[i]
print(f'\n\n\nThe smallest number in the random list, found using for loop is: {b}')


# Using np.min
print(f'\nThe smallest number in the random list, found using  np.min  is: {np.min(list_q_for)}')





'''
Finding how many number are greater than 40 AND summing them up
'''
c = 0
list_40_above = []
for i in range(q):
    if list_q_for[i] > 40:
        pass
        c += 1
        list_40_above.append(list_q_for[i])
print(f'\n\n\nThe quantity of numbers greater than 40 is: {c}')
print(f'\n\n\nThe  list  of  numbers  greater than 40 is: \n\n{list_40_above}')
print(f'\n\n\nThe sum of all  numbers greater than 40 is: {np.sum(list_40_above)}')
```
